```python
def MachCullochPitts (x,threshold):
  decision = None
  sum=0
  n=len(x)
  for i in range (0,n):
    sum+=x[i]
  if (sum>=threshold):
    decision=1
  else :
    decision=0
  return decision
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
```

```python
print ("-> AND")
threshold= 2 #
for o in X :
  print(o,MachCullochPitts(o,threshold))
```

```python
print ("-> OR")
threshold= 1 #
for o in X :
  print(o,MachCullochPitts(o,threshold))
```
